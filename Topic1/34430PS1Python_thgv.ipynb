## Constructing Employer-Employee Matched Data

```python
param = dict()
param['nk'] = 30 # Number of firm types
param['nl'] = 10 # Numer of worker types

param['α_σ'], param['ψ_σ'] = 1, 1 # Volatility of worker and firm effects

param['λ'] = 0.05 # Probability of moving firms

param['csort'] = 0.5 # Sorting effect
param['cnetw'] = 0.2 # Network effect
param['csig']  = 0.5 

param['w_σ'] = 0.2 # Volatility of wage shocks

param['num_time'] = 5   # Number of time periods
param['num_ind'] = 40000  # Number of individuals
param['firm_size'] = 15  # Max number of individuals per firm
```

```python
def gen_fixed_effects(param):
    
    nk, nl, α_σ, ψ_σ = param['nk'], param['nl'], param['α_σ'], param['ψ_σ']
    csort, cnetw, csig = param['csort'], param['cnetw'], param['csig'] 

    # Drawing fixed effects
    ψ = norm.ppf(np.linspace(1,nk,nk)/(nk+1)) * ψ_σ
    α = norm.ppf(np.linspace(1,nl,nl)/(nl+1)) * α_σ

    # Generating transition matrices.
    G = norm.pdf((ψ[ax, ax, :] - cnetw * ψ[ax, :, ax] - csort * α[:, ax, ax])/csig)
    G = np.divide(G, G.sum(axis=2)[:, :, ax])

    H = np.ones((nl,nk))/nl
    vec = np.zeros(nk)
    vec[0] = 1
    M = np.zeros((nk,nk))

    for l in range(0,nl):
        S, U = eig(G[l, :, :].T)
        stationary = np.array(U[:, np.where(np.abs(S-1.) < 1e-8)[0][0]].flat)
        stationary = stationary / np.sum(stationary)
        H[l, :] = stationary
        
    return ψ, α, G, H
```

### Simulating a Network

```python
def simulate_network(param):  

    # Generate fixed effects and transition matrices
    
    ψ, α, G, H = gen_fixed_effects(param)
    
    num_ind, num_time, firm_size = param['num_ind'], param['num_time'], param['firm_size']
    nk, nl, α_σ, ψ_σ, w_σ = param['nk'], param['nl'], param['α_σ'], param['ψ_σ'], param['w_σ'] 
    csort, cnetw, csig, λ = param['csort'], param['cnetw'], param['csig'],  param['λ'] 
    
    within_firm_count = num_ind/(firm_size*nk*num_time)

    network = np.zeros((num_ind,num_time), dtype=int)
    spellcount = np.ones((num_ind,num_time))

    # Random draws of worker types for all individuals in panel
    sim_worker_types = randint(low=1, high=nl, size=num_ind)

    for i in range(0, num_ind):

        l = sim_worker_types[i]
        # At time 1, we draw from H for initial firm
        network[i,0] = choices(range(0,nk),H[l,:])[0]

        for t in range(1,num_time):

            # Hit moving shock
            if rand() < λ:
                network[i,t] = choices(range(0,nk), G[l,network[i,t-1],:])[0]
                spellcount[i,t] = spellcount[i,t-1]+1
            else:
                network[i,t] = network[i,t-1]
                spellcount[i,t] = spellcount[i,t-1]      
                
    # Compiling IDs and timestamps     
    ids = np.reshape(np.outer(range(1,num_ind+1),np.ones(num_time)),(num_time*num_ind,1))
    ids = ids.astype(int)[:,0]
    ts = np.reshape(repmat(range(1,num_time+1),num_ind,1),(num_time*num_ind,1))
    ts = ts.astype(int)[:,0]

    # Compiling worker types
    types = np.reshape(np.outer(sim_worker_types, np.ones(num_time)),(num_time*num_ind,1))
    α_data = α[types.astype(int)][:,0]

    # Compiling firm types
    ψ_data = ψ[np.reshape(network,(num_time*num_ind,1))][:,0]
    k_data = np.reshape(network,(num_time*num_ind,1))[:,0]

    # Compiling spell data
    spell_data = np.reshape(spellcount,(num_time*num_ind,1))[:,0]
    
    # Merging all columns into a dataframe
    data = pd.DataFrame(data={'w_id':ids,'t':ts, 'k': k_data,
                              'α':α_data,'ψ':ψ_data,
                              'spell':spell_data.astype(int)})
    
    # Generating firm IDs
    def draw_fid(freq):

        max_int = np.int(np.maximum(1, freq.sum()/(firm_size * num_time)))
        return np.array(np.random.choice(max_int, size=freq.count()) + 1)
    
    dspell = data.groupby(['w_id', 'spell', 'k']).size().to_frame(name='freq').reset_index()
    dspell['f_id'] = dspell.groupby(['k'])['freq'].transform(draw_fid)
    dspell['f_id'] = dspell.groupby(['k', 'f_id'])['freq'].ngroup()
    dspell['f_id'] = dspell['f_id'] + 1
    
    data = data.merge(dspell, on=['w_id','spell', 'k'])
    
    data['move'] = (data['f_id'] != data['f_id'].shift(+1)) & (data['w_id'] == data['w_id'].shift(+1))
    
    # Computing wages through the AKM formula
    data['wage'] = data['α'] + data['ψ'] + w_σ * norm.rvs(size=num_ind * num_time)
    
    return data
```

# Question 6

For an example of an economic mechanism that cannot be captured in the basic AKM framework consider complementarities between unobservable skills that workers have. If, for example, higher skill workers are more productive in firms that have a high proportion of other high skill workers then we would expect to see this reflected in their wages. In this case we would expect to have an interaction term between the worker's skill level $\alpha_i$ and the $\alpha_{\tilde{i}}$ of other individuals in the firm. For a simple example it could depend on the interaction between average worker skill in the firm and a worker's own skill level. Letting $I(i,t)=\{l\neq i:j(l,t)=j(i,t)$ we could then have:

$$y_{it} = \alpha_i + \psi_{j(i,t)} + \alpha_i\left(\frac{1}{|I(i,t)|}\sum_{l\in I(i,t)} \alpha_l\right)+\varepsilon_{it}$$

With these kinds of spillovers between workers the identification of worker and firm fixed effects becomes more complicated and does not follow in the same way as in AKM. Some results may be possible though, in the vein of the work on social interactions.
